### Install dependencies from requirements.txt files
Use the shellscript below. Should be enough to run once.

```python
%%bash

PROJECT_DIR="/Users/Petra/Documents/repos/music_study_apr2024"
# Define the path to the requirements.txt file
REQUIREMENTS_FILE="$PROJECT_DIR/Code/requirements.txt"

# Attempt to install each package listed in requirements.txt
echo "Attempting to install packages from $REQUIREMENTS_FILE"

# Read each line (package) from the requirements file and install it
while IFS= read -r PACKAGE; do
    echo "Attempting to install package: $PACKAGE"
    # Try installing via conda first
    if conda install --yes $PACKAGE; then
        echo "Installation of $PACKAGE via conda successful!"
    else
        # If conda installation fails, try installing via pip
        echo "Package $PACKAGE not found on conda"
    fi
done < "$REQUIREMENTS_FILE"
```
